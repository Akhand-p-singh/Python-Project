```python
data['Rating'] = data['Rating'].fillna(data['Rating'].mean())
```

```python
data['Android Ver'].fillna(method = 'ffill', inplace = True)
```

```python
data['Size']  = data['Size'].apply(lambda x : x.strip('M'))
```

```python
data['Installs'] = data['Installs'].apply(lambda x : x.strip('+'))
```

```python
data['Installs'] = data['Installs'].apply(lambda x : x.replace(',', ''))
```

```python
data['Price'] = data['Price'].apply(lambda x : x.strip('$'))
```

```python
data['Reviews'] =pd.to_numeric(data['Reviews'], errors='coerce')
```

```python
data['Size'] = data['Size'].apply(lambda x : x.replace('Varies with device', '0'))
```

## Which Category has most Install 

```python
Downloads = data.groupby('Category').sum().sort_values(by='Installs', ascending= False).drop('1.9')
warnings.filterwarnings('ignore')
Downloads
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x= 'Installs', y= Downloads.index,  data = Downloads, palette= 'Reds_d').set_title('Most Download by Category', fontsize = '20')
```

## Which Category is Most Expensive 

```python
most_expensive = data.groupby('Category').sum().sort_values(by = 'Price', ascending = False).drop('1.9')
warnings.filterwarnings('ignore')
most_expensive.head(1)
```

```python
plt.figure(figsize = (15,10))
sns.barplot(y = most_expensive.index, x = most_expensive.Price, data = most_expensive ).set_title('Most Expensive App by Category', fontsize = '20')
```
